```python
# Histogram
'''Histograms represent the data distribution by forming bins along the range of the data and
then drawing bars to show the number of observations that fall in each bin.'''

import pandas as pd
import seaborn as sns
 
df = sns.load_dataset('iris')
sns.distplot(df['petal_length'],kde = True)
plt.show()
```

```python
# What is KDE?
Kernel Density Estimation (KDE) helps in estimating the probability density function of a continuous random
variable. It is used for non-parametric analysis.
Non parametric do not assume that the data is normally distributed.
The major/most likey non parametric test you are likely to ever use is the chi-square test.

hist=False will show the kde plot

Refer: https://www.statisticshowto.datasciencecentral.com/parametric-and-non-parametric-data/ for more details
```

```python
import pandas as pd
import seaborn as sns
 
df = sns.load_dataset('iris')
sns.distplot(df['petal_length'],hist=False)
plt.show()
```

```python
#If you want to view parametric dist then use distplot

df = sns.load_dataset('iris')
sns.distplot(df['petal_length'])
plt.show()
```

```python
# As discussed in previous class about univariate, bivariate
#Bivariate distribution helps in determining relationship between two variables- jointplot() helps here
"""
Incase where data is dense we use scatterplot
Incase if data is sparse ie. scattered we make use of hexbinplot
"""

sns.jointplot(x = 'petal_length',y = 'petal_width',data = df)
plt.show()

sb.jointplot(x = 'petal_length',y = 'petal_width',data = df,kind = 'hex')
plt.show()
```

```python
# In case of multivariate data where we wish to visualize relationships between indivual variables
#we perfer to use pair-plots

sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

```python
""" Swarm Plot
This gives a better representation of the distribution of values, but it does not scale well to 
large numbers of observations.
This style of plot is sometimes called a “beeswarm”.
"""

import pandas as pd
import seaborn as sns
 
df = sns.load_dataset('iris')
sns.swarmplot(x = "species", y = "petal_length", data = df)
plt.show()
```

```python
'''
source:wikipedia
A violin plot is a method of plotting numeric data.
It is similar to a box plot, with the addition of a rotated kernel density plot on each side.
Violin plots are similar to box plots, except that they also show the probability density of the data
at different values, usually smoothed by a kernel density estimator.
'''

import pandas as pd
import seaborn as sb
 
#df = sb.load_dataset('tips')
#sb.violinplot(x = "day", y = "total_bill", data=df)
df = sb.load_dataset('iris')
sb.violinplot(x = "species", y = "petal_length", data = df)
plt.show()
```
